```python
import pandas as pd

from ggplot import *
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('./hr_year.csv',usecols=['yearID','HR'])
df.head()
```

```python
ggplot(df,aes('yearID','HR'))+geom_point(aes(color='red'))+geom_line(aes(color='red'))
```
