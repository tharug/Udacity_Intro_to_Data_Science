```python
# creating a function to calculate number of rainy days
import pandasql

def num_rainy_days(filename):
    weather_data=pd.read_csv(filename)
      
    # using pandasql to execute a sql query on the pandas dataframe
    
    # defining the query:q
    q="select sum(rain) from weather_data"
    
    # Execute the SQL command against the dataframe
    rainy_days=pandasql.sqldf(q,locals())
    
    return rainy_days
```

```python
rainy_days=num_rainy_days(filename)
rainy_days
```

```python
# creating a function to identify the max temperature for fog

def max_temp_aggregate_by_fog(filename):
    weather_data=pd.read_csv(filename)
      
    # using pandasql to execute a sql query on the pandas dataframe
    
    # defining the query:q
    q="select fog, max(maxtempi) from weather_data group by fog"
    
    # Execute the SQL command against the dataframe
    foggy_days=pandasql.sqldf(q,locals())
    
    return foggy_days
```

```python
max_temp_aggregate_by_fog(filename)
```

```python
import csv
```

```python
f_in=open('./turnstile-110528.txt','r')
f_out=open('./updated_'+'turnstile-110528.csv','w')
```

```python
reader=csv.reader(f_in,delimiter=',')
writer=csv.writer(f_out,delimiter=',')
```

```python
for line in reader:
    row_lenght=len(line)
    rec_lenght=5
    prefix=line[:3]
    for record in range(3,row_lenght,rec_lenght):
        line_out=prefix+line[record:record+5]
        writer.writerow(line_out)
```

```python
df=pd.read_csv('./updated_turnstile-110528.csv',header=None,
               names=['C/A','UNIT','SCP','DATEn','TIMEn','DESCn','ENTRIESn','EXITSn'])
```

```python
# Creating column to identify the hourly enteries
# Calculated by getting the difference for each ENTRIESn row with the row above
df['ENTRIESn_hourly']=(df['ENTRIESn']-df['ENTRIESn'].shift(1)).fillna(1)
```

```python
# Creating column to identify the hourly exits
# Calculated by getting the difference for each exitsn row with the row above
df['EXITSn_hourly']=(df['EXITSn']-df['EXITSn'].shift(1)).fillna(0)
```
